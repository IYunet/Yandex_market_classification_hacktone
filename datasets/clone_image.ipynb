```python
#!c1.8
import pandas as pd
import os
import requests
from PIL import Image
from io import BytesIO
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

input_file = "/home/jupyter/work/resources/data.csv"
df = pd.read_csv(input_file).drop(columns=["market_sku_id"])
df.head()
```

```python
#!c1.8
#pragma dataset init pictures --size 150Gb

def download_pic(idx, link, verdict):
    try:
        # URL изображения для загрузки
        image_url = "https://" + link

        # Загружаем изображение
        response = requests.get(image_url)
        if response.status_code == 200:
            image_data = BytesIO(response.content)
            img = Image.open(image_data)
            img = img.convert("RGB")
            save_path = os.path.join("/home/jupyter/mnt/datasets/pictures", "pic_" + str(idx) + ".jpg")
            img.save(save_path)
    except Exception as e:
        pass

for i in tqdm(range(df.shape[0])):
    row = df.loc[i]
    new_df = download_pic(i, row[0], row[1])
```
